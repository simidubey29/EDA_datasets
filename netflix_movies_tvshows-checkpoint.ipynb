```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv(r'C:\Users\panka\Downloads\netflix_titles.csv')

df=df.dropna(subset=['type','release_year','rating','country','duration'])

type_counts=df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=['skyblue','red'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('movies_vs_tvshows.png')
plt.show()


```

```python

rating_counts=df['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_counts,labels=rating_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('percentage of content ratings')
plt.tight_layout()
plt.savefig('content_rating_pie.png')
plt.show()
```

```python
movie_df=df[df['type']=='Movie'].copy()
movie_df['duration_int']=movie_df['duration'].str.replace('min','').astype(int)

plt.figure(figsize=(8,6))
plt.hist(movie_df['duration_int'],bins=30, color='purple',edgecolor='blue')
plt.title('distribution of movie duration')
plt.xlabel('duration(min)')
plt.ylabel('no.of movies')
plt.tight_layout()
plt.savefig('movies_duration_hist.png')
plt.show()
```

```python
release_counts=df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index,release_counts.values,color='green')
plt.title('release_yr vs no.of shows')
plt.xlabel('release_yr')
plt.ylabel('no.of shows')
plt.tight_layout()
plt.savefig('release_yr_scattter.png')
plt.show()
```

```python
country_counts=df['country'].value_counts().head(10)
plt.figure(figsize=(9,7))
plt.bar(country_counts.index,country_counts.values,color='purple')
plt.title('top 10 country by no_of_shows')
plt.ylabel('country')
plt.xlabel('no.of shows')
plt.tight_layout()
plt.savefig('top10_countries.png')
plt.show()
```

```python
content_by_yr =df.groupby(['release_year','type']).size().unstack().fillna(0)

fig, ax=plt.subplots(1,2,figsize=(12,5))
#first subplot:movies
ax[0].plot(content_by_yr.index,content_by_yr['Movie'],color='blue')
ax[0].set_title('movie release per yr')
ax[0].set_xlabel('year')
ax[0].set_ylabel('number of movies')

#second subplot:tv shows
ax[0].plot(content_by_yr.index,content_by_yr['TV Show'],color='red')
ax[0].set_title('TV shows release per yr')
ax[0].set_xlabel('year')
ax[0].set_ylabel('number of movies')

fig.suptitle('comparison of movies and TV shows relased over yr')
plt.tight_layout()
plt.savefig('movies_tv_shows_compare.png')
plt.show()
```
